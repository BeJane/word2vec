```python
from gensim.models import KeyedVectors
import numpy as np
import pprint
```

```python
path = 'renminribao/embeddings_9epoch_81.413loss.txt'
word2vec = KeyedVectors.load_word2vec_format(path,binary=False,unicode_errors='ignore')
```

```python
words = ['新冠','人民','发展','绿色','北京','先生','数字','产业','经济']
for w in words:
    print('words: {}'.format(w))
    print('similar words:')
    pprint.pprint(word2vec.most_similar(w))
```

```python
w1 = ['男','女','城市','经济']
w2 = ['博士','女士','建设','发展']
w3 = ['女','男','农村','生态环境']
for a,b,c in zip(w1,w2,w3):
    print('{}--{},\n{}--?'.format(a,b,c))
    pprint.pprint(word2vec.most_similar(positive=[a,b],negative=[c]))
```
